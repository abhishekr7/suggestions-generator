# Import necessary libraries

```python
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Business dataset 

```python
article_path = '/home/abhishek/Documents/NLP/Blog/suggestions-generator/bbc-news-summary/articles/sport/'
summary_path = '/home/abhishek/Documents/NLP/Blog/suggestions-generator/bbc-news-summary/summaries/sport/'
n_examples = 511
```

# Create a word list for both articles and summaries

```python
Tokenizer = RegexpTokenizer(r'[A-Za-z0-9\$]+')
article_word_list = []
summary_word_list = []

for i in range(1,n_examples+1):
    
    # ARTICLE
    
    with open(article_path+'{:03}.txt'.format(i),'r') as article_file:
        text = article_file.read()
        word_list = [word.lower() for word in Tokenizer.tokenize(text)]
        article_word_list.extend(word_list)
        
    # SUMMARY
    
    with open(summary_path+'{:03}.txt'.format(i),'r') as summary_file:
        text = summary_file.read()
        word_list = [word.lower() for word in Tokenizer.tokenize(text)]
        summary_word_list.extend(word_list)
```

# Create a word set for both articles and summaries

```python
article_word_set = set(article_word_list)
summary_word_set = set(summary_word_list)
```

# Create a common word set

```python
all_word_set = set().union(article_word_set).union(summary_word_set)
```

```python
len(all_word_set)
```

# number of times "only" used in article 

```python
def when_in_article_only(WORD):
    
    internal_count = 0
    count = 0
    
    for i in range(1,n_examples+1):
        
        in_article = False
        in_summary = False
        
        # ARTICLE
        
        with open(article_path+'{:03}.txt'.format(i),'r') as article_file:
            text = article_file.read()
            
            word_list = [word.lower() for word in Tokenizer.tokenize(text)]
            
            if WORD in word_list:
                in_article = True
    
                for x in word_list:
                    if x == WORD:
                        internal_count += 1
                    
        
        # SUMMARY
        
        with open(summary_path+'{:03}.txt'.format(i),'r') as summary_file:
            text = summary_file.read()
            
            word_list = [word.lower() for word in Tokenizer.tokenize(text)]
            
            if WORD in word_list:
                in_summary = True
        
        if in_article == True and in_summary == False:
            count += internal_count
        
    return count
```

# number of times "only" used in summary 

```python
def when_in_summary_only(WORD):
    
    internal_count = 0
    count = 0
    
    for i in range(1,n_examples+1):
        
        in_article = False
        in_summary = False
        
        # ARTICLE
        
        with open(article_path+'{:03}.txt'.format(i),'r') as article_file:
            text = article_file.read()
            
            word_list = [word.lower() for word in Tokenizer.tokenize(text)]
            
            if WORD in word_list:
                in_article = True
                    
        
        # SUMMARY
        
        with open(summary_path+'{:03}.txt'.format(i),'r') as summary_file:
            text = summary_file.read()
            
            word_list = [word.lower() for word in Tokenizer.tokenize(text)]
            
            if WORD in word_list:
                in_summary = True
                
                for x in word_list:
                    if x == WORD:
                        internal_count += 1
        
        if in_article == True and in_summary == False:
            count += internal_count
        
    return count
```

# number of times used in both

```python
def when_in_both(WORD):
    
    internal_count_article = 0
    internal_count_summary = 0
    
    count = 0
    
    for i in range(1,n_examples+1):
        
        in_article = False
        in_summary = False
        
        # ARTICLE
        
        with open(article_path+'{:03}.txt'.format(i),'r') as article_file:
            text = article_file.read()
            
            word_list = [word.lower() for word in Tokenizer.tokenize(text)]
            
            if WORD in word_list:
                in_article = True
                
                for x in word_list:
                    if x == WORD:
                        internal_count_article += 1
        
        # SUMMARY
        
        with open(summary_path+'{:03}.txt'.format(i),'r') as summary_file:
            text = summary_file.read()
            
            word_list = [word.lower() for word in Tokenizer.tokenize(text)]
            
            if WORD in word_list:
                in_summary = True
                
                for x in word_list:
                    if x == WORD:
                        internal_count_summary += 1
        
        if in_article == True and in_summary == True:
            count += min(internal_count_article,internal_count_summary)
        
    return count   
```

# total number of times used

```python
def total_usage(word):
    count = 0
    
    for item in article_word_list:
        if item == word:
            count += 1
    
    for item in summary_word_list:
        if item == word:
            count += 1
    
    return count 
```

# create a dataframe with all usage counts

```python
df = pd.DataFrame(columns=['word','#article','#summary','#both','#total'])
df.set_index(['word'])

cnt = 0

for item in all_word_set:
    cnt += 1
    w = when_in_article_only(item)
    x = when_in_summary_only(item)
    y = when_in_both(item)
    z = total_usage(item)
    df.loc[item] = [item,w,x,y,z]
    print(cnt)
```

# Find relative rates of usage

```python
df['R_article'] = df['#article']/df['#total']
```

```python
df['R_summary'] = df['#summary']/df['#total']
```

```python
df['R_both'] = df['#both']/df['#total']
```

```python
df
```

```python
df['x_not'] = df['R_both'] - df['R_article']
```

```python
df['y_not'] = df['R_both'] - df['R_summary']
```

```python
df
```

```python
stop_words = set(stopwords.words('english'))

word_label = df['word'].tolist()
x_coord = df['x_not'].tolist()
y_coord = df['y_not'].tolist()

area = np.pi*0.1

plotted = set()

for i,label in enumerate(word_label):
    #print(i)
    
    if label in stop_words:
        continue
    
    temp_x = x_coord[i]
    temp_y = y_coord[i]
    
    #print(str(temp_x) + " " + str(temp_y))
    
    x = abs((temp_x - temp_y - 1)/np.sqrt(2))
    y = abs((temp_x + temp_y + 1)/np.sqrt(2))
    
    #print(x)
    
    if (x,y) not in plotted:
        plt.scatter(x,y, s=area, color = 'red')
        x_label = x + 0.005
        y_label = y + 0.005
        
        #if y_label > 1.4:
            #plt.text(x_label, y_label, label, fontsize=9)
        plotted.add((x,y))
    #print(label)
plt.show()
```

```python
df['x_coord'] = abs((df['x_not'] - df['y_not'] - 1)/np.sqrt(2))
```

```python
df['y_coord'] = abs((df['x_not'] + df['y_not'] + 1)/np.sqrt(2))
```
